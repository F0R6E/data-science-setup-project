```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

##### Pandas

```python
df = pd.read_csv(
    'data/btc.csv',
    header=None,
    names=['Timestamp', 'Price'],
    index_col=0,
    parse_dates=True
)
```

```python
df.head()
```

###### Matplotlib

```python
df.plot(figsize=(15, 7))
```

###### Seaborn

```python
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize=(14, 8))
sns.distplot(df['Price'], ax=ax)
```

###### Bokeh

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import Span
output_notebook()
```

```python
p = figure(
    title="Bitcoin Price", x_axis_label='Time', y_axis_label='Price', width=940, height=550,
    x_axis_type="datetime"
)
p.line(df.index, df['Price'], legend="Price", line_width=1, line_color='green')

show(p)
```
